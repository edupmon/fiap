```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('dados/card_transdata.tar.gz')
```

```python
X = df.drop(columns=['fraud'])
y = df['fraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    stratify=y, random_state=7)
```

```python
rf = RandomForestClassifier(n_estimators=5, max_depth=3, random_state=7)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
def print_tree(tree):
    fig, ax = plt.subplots(figsize=(12, 12))
    plot_tree(rf.estimators_[tree],
              feature_names=df.drop(columns=['fraud']).columns.to_list(),
              class_names=['fraud', 'ok'],
              filled=True)
    plt.show()
```

```python
for i in range(0, len(rf.estimators_)):
    print(rf.estimators_[i])
    print_tree(i)
```

```python
rf.score(X_train, y_train)
```
