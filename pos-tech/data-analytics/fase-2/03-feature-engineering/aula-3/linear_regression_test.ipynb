```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
# Generate sample data
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Feature
y = 2 * X + np.random.randn(100, 1)  # Target with some noise
```

```python
# Split data into training and testing sets
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]
```

```python
# Linear Regression with raw data
model_raw = LinearRegression()
model_raw.fit(X_train, y_train)
y_pred_raw = model_raw.predict(X_test)
```

```python
# MinMax Scaler
scaler_minmax = MinMaxScaler()
X_train_minmax = scaler_minmax.fit_transform(X_train)
X_test_minmax = scaler_minmax.transform(X_test)
```

```python
# Linear Regression with MinMax scaled data
model_minmax = LinearRegression()
model_minmax.fit(X_train_minmax, y_train)
y_pred_minmax = model_minmax.predict(X_test_minmax)
```

```python
# Standard Scaler
scaler_standard = StandardScaler()
X_train_standard = scaler_standard.fit_transform(X_train)
X_test_standard = scaler_standard.transform(X_test)
```

```python
# Linear Regression with Standard scaled data
model_standard = LinearRegression()
model_standard.fit(X_train_standard, y_train)
y_pred_standard = model_standard.predict(X_test_standard)
```

```python
# Evaluate models
mse_raw = mean_squared_error(y_test, y_pred_raw)
mse_minmax = mean_squared_error(y_test, y_pred_minmax)
mse_standard = mean_squared_error(y_test, y_pred_standard)

print(f'MSE with Raw Data: {mse_raw}')
print(f'MSE with MinMax Scaled Data: {mse_minmax}')
print(f'MSE with Standard Scaled Data: {mse_standard}')
```

```python
# Plot results
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred_raw, color='blue')
plt.title('Raw Data')

plt.subplot(1, 3, 2)
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred_minmax, color='green')
plt.title('MinMax Scaled Data')

plt.subplot(1, 3, 3)
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred_standard, color='red')
plt.title('Standard Scaled Data')

plt.show()
```
