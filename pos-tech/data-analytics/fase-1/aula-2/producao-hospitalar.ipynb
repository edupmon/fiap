# Analisando dados de produção hospitalar

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
dados = pd.read_csv('dados/sih_cnv_qiuf074447104_28_63_99.csv',
                    engine = 'python', encoding = 'ISO-8859-1', skiprows = 3,
                    sep = ';', skipfooter = 12, thousands = '.', decimal = ',')
```

```python
dados[['Unidade da Federação', '2023/Dez']]
```

```python
axis = dados.plot(x = 'Unidade da Federação', y = '2023/Dez', kind = 'bar',
                  figsize = (12, 6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title('Valor por Unidade da Fedaração')
plt.xticks(rotation = 85)
plt.show()
```
