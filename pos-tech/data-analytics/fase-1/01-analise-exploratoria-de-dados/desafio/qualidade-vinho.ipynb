```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1 - Carregando os Dados

```python
dados = pd.read_csv('dados/winequalityN.csv')
```

# 2 - Estrutura dos Dados

```python
dados.head()
```

```python
dados.describe()
```

```python
dados.info()
```

# 3 - Removendo Nulos e Duplicados

```python
dados.dropna(inplace=True)
```

```python
dados.drop_duplicates(inplace=True)
```

# 4 - Análise dos Dados

## a - Distribuição da Acidez Volátil por Qualidade do Vinho

```python
sns.violinplot(x='quality', y='volatile acidity', data=dados)

plt.title('Distribuição da Acidez Volátil por Qualidade do Vinho')
plt.xlabel('Qualidade')
plt.ylabel('Acidez Volátil')

plt.show()
```

## b - Distribuição do Teor Alcóolico por Qualidade do Vinho

```python
sns.violinplot(x='quality', y='alcohol', data=dados)

plt.title('Distribuição do Teor Alcóolico por Qualidade do Vinho')
plt.xlabel('Qualidade')
plt.ylabel('Teor Alcóolico')

plt.show()
```

## c - Relação entre Acidez Volátil e Teor Alcóolico

```python
sns.scatterplot(x='volatile acidity', y='alcohol', data=dados)

plt.title('Relação entre Acidez Volátil e Teor Alcóolico')
plt.xlabel('Acidez Volátil')
plt.ylabel('Teor Alcóolico')

plt.show()
```

## d - Relação entre Teor Acóolico e Qualidade do Vinho

```python
sns.scatterplot(x='alcohol', y='quality', data=dados)

plt.title('Relação entre Teor Acóolico e Qualidade do Vinho')
plt.xlabel('Teor Alcóolico')
plt.ylabel('Qualidade')

plt.show()
```

# 5 - Resultados das Análises

## a - Qual a acidez volátil média dos vinhos de qualidade elevada?

```python
dados.groupby('quality').agg({'volatile acidity':'mean'})
```

```python
_ = dados.groupby('quality').agg({'volatile acidity':'mean'}).plot()
```

## b - Qual a relação entre teor alcólico e qualidade do vinho?

```python
dados.groupby('quality').agg({'alcohol':'mean'})
```

```python
_ = dados.groupby('quality').agg({'alcohol':'mean'}).plot()
```

## c - Há alguma relação entre acidez volátil e teor alcóolico?

Visualmente, não parece existir uma correlação entre acidez volátil e teor alcóolico, o que podemos conferir com o baixo índice de correlação entre as duas variáveis.

```python
dados[['volatile acidity', 'alcohol']].corr()
```

## d - Há alguma relação entre acidez volátil e qualidade do vinho?

Visualmente podemos observar uma fraca correlação entre as duas variáveis, a medida que a qualidade aumenta a acidez volátil diminui.

```python
sns.scatterplot(x='volatile acidity', y='quality', data=dados)

plt.title('Relação entre Acidez Volátil e Qualidade do Vinho')
plt.xlabel('Acidez Volátil')
plt.ylabel('Qualidade')

plt.show()
```

```python
dados[['volatile acidity', 'quality']].corr()
```
