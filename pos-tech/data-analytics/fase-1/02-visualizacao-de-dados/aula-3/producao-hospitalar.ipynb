```python
dados_sudeste = dados_sudeste / 1_000_000

ax = dados_sudeste.T.plot(figsize = (12, 6))
ax.set_ylabel('Milhões de Reais')

plt.show()
```

```python
# Gastos dos últimos 5 anos - Região Sudeste
dados_5_anos = dados_sudeste[dados_sudeste.columns[-61:]].T

# Tamanho do gráfico
ax = dados_5_anos.plot(figsize = (12, 6))

# Marcador do eixo x a cada ano
ax.set_xticks(range(0, 61, 12), dados_5_anos.index.to_list()[::12])

# Títulos dos eixos
ax.set_xlabel('Mês')
ax.set_ylabel('Milhões de Reais')

# Linha de grade
ax.grid(linestyle = '--')

plt.ylim(0, 600)
plt.show()
```

```python
# Dados do mês mais recente disponível
# (penúltima coluna do dataframe - a última columa é a região)

ult_mes = dados_analise.columns[-2]

dados_ult_mes = dados_analise[ult_mes]\
                .sort_values(ascending = False)\
                .copy()
dados_ult_mes.head()
```

```python
# Comparando São Paulo com os demais estados

compara_estados = dados_ult_mes / dados_ult_mes.loc['35 São Paulo']
compara_estados.head()
```

```python
ax = compara_estados.plot(kind = 'bar', figsize = (12, 6))

ax.set_ylabel('Comparção de Gastos com o Estado de SP')
ax.grid(linestyle = '--')
ax.patches[0].set_facecolor('orange')
```

```python
# Preparando para cruzar os dados do último mês com a população
# (ajustando o índice)

df_ult_mes = pd.DataFrame(dados_ult_mes)
df_ult_mes.index = [estado[3:] for estado in dados_ult_mes.index.to_list()]
df_ult_mes.head()
```

```python
# Carregando dados da população (obtidos através de um prompt no ChatGPT)

df_populacao = pd.read_csv('dados/populacao.csv')
df_populacao.set_index('Unidade da Federação', inplace = True)
df_populacao.head()
```

```python
# Juntando os dados de gastos com os dados da população
df_ult_mes_com_pop = df_ult_mes.join(df_populacao)

# Coluna de gastos por pessoa
df_ult_mes_com_pop['2023/Dez por Pessoa'] = \
    df_ult_mes_com_pop['2023/Dez'] / df_ult_mes_com_pop['População']
```

```python
# Comparando gastos totais dos últimos dois anos

# Último ano
gastos_2023 = pd.DataFrame(dados_analise[dados_analise.columns[-13:-1]]\
                           .sum(axis = 1))
gastos_2023.rename(columns = {0: 'Gastos 2023'}, inplace = True)

# Penúltimo ano
gastos_2022 = pd.DataFrame(dados_analise[dados_analise.columns[-25:-13]]\
                           .sum(axis = 1))
gastos_2022.rename(columns = {0: 'Gastos 2022'}, inplace = True)

# Gastos dos dois anos
gastos_2_anos = gastos_2022.join(gastos_2023)
gastos_2_anos.index = [estado[3:] for estado in gastos_2_anos.index.to_list()]

# Incluindo a população
gastos_2_anos = gastos_2_anos.join(df_populacao)
gastos_2_anos.head()
```

```python
# Ajustando o dataframe para análise da evolução mensal de gastos

analise_mensal = dados_analise.drop(columns = ['Região']).T.copy()
```

```python
# Gráfico da evolução mensal de gastos

plt.figure(figsize = (12, 6))

ax = sns.lineplot(data = analise_mensal, x = analise_mensal.index,
                  y = '31 Minas Gerais', label = 'MG')
ax = sns.lineplot(data = analise_mensal, x = analise_mensal.index,
                  y = '32 Espírito Santo', label = 'ES')
ax = sns.lineplot(data = analise_mensal, x = analise_mensal.index,
                  y = '33 Rio de Janeiro', label = 'RJ')
ax = sns.lineplot(data = analise_mensal, x = analise_mensal.index,
                  y = '35 São Paulo', label = 'SP')

ax.xaxis.set_major_locator(ticker.IndexLocator(base = 12, offset = 0))
ax.xaxis.set_minor_locator(ticker.IndexLocator(base = 12, offset = 6))
ax.xaxis.grid(True, which='both')

plt.xticks(rotation = 30)
plt.ylim(0, math.ceil(max(analise_mensal.max()) * 1.1))
plt.grid(linestyle = '--')
plt.ylabel('Estado')
plt.xlabel('Mês')

plt.show()
```
